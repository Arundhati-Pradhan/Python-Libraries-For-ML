# TASK 6 : Prediction using Decision Tree Algorithm
Create the Decision Tree classifier and visualize it graphically.The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

---



```python
# import libraries
import pandas as pd      #for loading dataset
import numpy as np       #for numerical computation
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
# import dataset
from sklearn.datasets import load_iris
iris=load_iris()  
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
X=iris.data     
Y=iris.target   
```

# Splitting dataset into Train and Test sets

---



```python
X_train,X_test, y_train, y_test=train_test_split(X,Y, test_size=0.20)  
```

# Import the DecisionTreeClassifier object and fit the model

---



```python
tree_clf=DecisionTreeClassifier()     
tree_clf.fit(X_train, y_train)      # fit the model

print('Decision Tree Classifer Created Successfully')
```

```python
#prediction
y_pred=tree_clf.predict(X_test)   
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn import metrics
#actual = y_test
accuracy=metrics.accuracy_score(y_test,y_pred)
accuracy
```

# Constructing Decision Tree

```python
import matplotlib.pyplot as plt
from sklearn import tree


a=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
b=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(tree_clf, feature_names = a, class_names = b, filled = True);
```
